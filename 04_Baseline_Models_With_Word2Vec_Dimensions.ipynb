```python
'''

This script will create Word2Vec dimesnions for every row in our dataset and then run a few baseline models.   

1) First we create word vectors for every row, then just average those word vectors to create sentence vectors.
2) Apply baseline models using following scoring metric and plot them for comparison
   -accuracy
   -recall  (positive class i.e jokes important not to miss)

'''

#Import required libraries
import pandas as pd
from gensim.models import Word2Vec

from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics

#Visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
from pandas.tools.plotting import scatter_matrix

#Configure Visualization Defaults
#%matplotlib inline = show plots in Jupyter Notebook browser
%matplotlib inline
mpl.style.use('ggplot')
sns.set_style('white')
pylab.rcParams['figure.figsize'] = 12,8
```

```python
#Read created input data
df=pd.read_csv('datasets/Final_data.csv')

print(len(df))
```

```python
df=df[~df.sentence.isnull()]
```

```python
#Create word vectors for every row

text=[]
for index,row in df.iterrows():
  print(str(row['sentence'])+"\n")
  sentence_word_list = [word for word in str(row['sentence']).lower().split()]
  text.append(sentence_word_list)
w2v = Word2Vec(text, min_count=1)
```

```python
#Create sentence vectors. For now, we'll just do this by averaging word vectors created in step before 
vector_record=[]
for index,row in df.iterrows():
  sentence = str(row['sentence'])
  if len(sentence)!=0:
    sentence_vect = sum([w2v[w] for w in sentence.lower().split()])/(len(sentence.split())+0.001)
  else:  
    sentence_vect = np.zeros((100,)) #shouldn't come here ideally
  vector_record.append(sentence_vect)    
```

```python
#Chose the first 100 dimensions for modelling
X = pd.DataFrame(vector_record, columns=range(100))
print(X.head())
print(X.shape)


```

```python
y = df['class']
print(type(y))
print(y.shape)
```

```python
#Confirgure for modelling
seed = 23
#Generate model objects
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
#Evaluate each model & store results
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
#Create a chart comparing algorithms based on accuracy metric
fig = plt.figure()
fig.suptitle('Algorithm Comparison (** based on accuracy**)')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
